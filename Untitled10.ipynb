## Import bibliothèque

```python
import random
import numpy as np
```

## Variables utiles et modifiables

```python
# Variable
#Variables du système

# Nombre de clients maximum dans le système
B = 6
# Nombre de serveurs maximum dans le système
K = 2 
# Nombre maximum de clients dans un batch
b = 3
# Débit des serveurs par slots
d = 1


#Variables de coûts 

# coût d’une VM en fonctionnement par slot
Cf = 1
# coût d’activation d’une VM
Ca = 1
# coût de désactivation d’une VM
Cd = 1
# coût de maintien d’un client dans la file
Ch = 1

# Entrainement 
episodes = 10
max_steps = 10000
learning_rate = 0.1
alpha = 1 # facteur d'apprentissage
gamma = 0.9 # facteur d'actualisation
epsilon = 0.9
```

## Class environnement

```python
class Env():
    def __init__(self):
        self.vm_actif = 0
        self.personne_en_file = 0
        self.cout = 0
        self.arrivee = 0
    
    def step(self, action: int) -> (int, int): # retoure observation, cout
        self.arrivee = self.generer_arrive()
        self.personne_en_file = min(B,max(0,self.personne_en_file + self.arrivee - self.vm_actif * d ))
        self.cout = self.calc_cout(self.personne_en_file, action, self.vm_actif)
        self.vm_actif = action
        return (self.personne_en_file, self.vm_actif), self.cout

    def reset(self) -> (int, int):
        self.vm_actif = 0
        self.personne_en_file = 0
        self.cout = 0
        return (self.personne_en_file, self.vm_actif)

    def calc_cout(self,personne: int, action: int, vm_actif: int):
        c = vm_actif*Cf
        if action > vm_actif :
            c += (vm_actif - action)*Ca
        else :  
            c += (action - vm_actif)*Cd
        c += personne*Ch
        return c

    def generer_arrive(self) -> int:
        return random.randint(0,3)
```

## Test de l'environnement

```python
env = Env() 

for i in range(episodes):
    obs = env.reset()
    print("Episode {} commence à {}".format(i,obs))
    for j in range(max_steps):
        action = random.randint(0,3)
        (personne_en_file, vm_actif), cout = env.step(action)
        print("Arrivée : {}, Action choisie : {}, obs après : ({},{})".format(env.arrivee,action, personne_en_file, vm_actif))
```

## Qtable

```python
# Création du dictionnaire poour la matrice de coûts
def dictionnaire_indice(B,K):
    d_indice = []
    d_inv = {}
    
    c=0
    
    for i in range(B+1):
        for j in range(K+1):
            d_indice.append((i,j))
            d_inv[(i,j)]=c
            
            c+=1
    return d_indice,d_inv

_,dictio = dictionnaire_indice(B,K)
dictio
```

```python
q_table = np.zeros(((B+1)*(K+1),K+1))
q_table[dictio.get((0,1))]# chercher les q_valeur pour l'état (0,1) personne/vm_active
q_table
```

```python
env = Env() 

for i in range(episodes):
    obs = env.reset()
    print("Episode {} commence à {}".format(i,obs))
    for j in range(max_steps):
        # politique epsilon greedy
        p = np.random.random()
        if p < epsilon:
          action = np.argmax(q_table[dictio.get(obs)])
        else:
          action = random.randint(0,K)
      
        old_obs = obs 
        obs, cout = env.step(action)
        # print("Arrivée : {}, Action choisie : {}, obs après : ({},{})".format(env.arrivee,action, obs[0], obs[1]))

       
        q_value = q_table[dictio.get(old_obs)] [action-1] + gamma *(-cout + np.argmax(q_table[dictio.get(obs)]) - q_table[dictio.get(old_obs)] [action-1])   
        q_table[dictio.get(old_obs)] [action-1] = q_value 
```
